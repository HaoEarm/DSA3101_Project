```python
# the length of data changes everytime i run 
app_reviews = reviews_all('com.revolut.revolut',lang='en',country='sg')
data = app_reviews
revolut_play_df = pd.DataFrame.from_records(data)
revolut_play_df
```

```python
revolut_play_df['Bank'] = 'Revolut'
revolut_play_df = revolut_play_df.reindex(columns=['userName', 'content', 'score', 'at', 'Bank'])
revolut_play_df.columns = ['UserName', 'Review', 'Score', 'Date', 'Bank']
revolut_play_df
```

```python
def convert_emojis_to_words(text):
    converted_text = emoji.demojize(text)
    return converted_text

revolut_play_df['Review'] = revolut_play_df['Review'].apply(convert_emojis_to_words)
revolut_play_df
```
